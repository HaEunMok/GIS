```python
m=folium.Map(
  location=[33.3684955195788, 126.52918183373025],    # 한라산 위경도
  zoom_start =10,
  #tiles ='Stamen Terrain' # 레이아웃형태
)

tooltip ='Click!!'

folium.Marker(
    [33.3684955195788, 126.52918183373025],
    popup='<strong>한라산</strong>', # marker 클릭시 팝업창 내용
    tooltip =tooltip,
    icon=folium.Icon(color='#ff0000', icon='info-sign')
).add_to(m)
folium.Marker(
    [33.506424, 126.492089],
    popup='<strong>제주국제공항</strong>', # marker 클릭시 팝업창 내용
    tooltip =tooltip,
    icon=folium.Icon(color='green', icon='info-sign')
).add_to(m)
folium.Marker(
    [33.508461049711265, 126.54097984346038],
    popup='<strong>바울랩</strong>', # marker 클릭시 팝업창 내용
    tooltip =tooltip,
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(m)
m
```

```python
오름['이미지']=오름['오름명']+'.jpg'
```

```python
import folium

m = folium.Map(
  location=[33.3684955195788, 126.52918183373025],    # 한라산 위경도
  zoom_start=10,
  #tiles ='Stamen Terrain' # 레이아웃형태
)

tooltip = 'Click!!'

for i in range(오름.shape[0]):
  folium.Marker(
    [오름.iloc[i]['위도'], 오름.iloc[i]['경도']],
    popup=f'<div style="width:100px"><strong>{오름.iloc[i]["오름명"]}</strong><br>\
    주차장 : {오름.iloc[i]["주차장"]}<br>\
    화장실 : {오름.iloc[i]["화장실"]}<br>\
    <img width="80px" src="jeju-74956_1280.jpg"><br>\
    <a href="https://visitjeju.net/kr/search?q=%EC%98%A4%EB%A6%84">상세페이지 이동</a></div>',
    tooltip=tooltip
  ).add_to(m)

m.save('map.html')  # 지도를 HTML 파일로 저장
```
