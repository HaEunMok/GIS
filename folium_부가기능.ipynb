```python
from haversine import haversine
import folium

# haversine(위치1, 위치2, unit ='km')

m = folium.Map(
  location=[33.3684955195788, 126.52918183373025],    # 한라산 위경도
  zoom_start=10,
  #tiles ='Stamen Terrain' # 레이아웃형태
)

tooltip = 'Click!!'

img= "https://drive.google.com/file/d/1DjqbQe2dE4OWqKTz7Gx1Bp_-RCnM23up/view?usp=sharing"

for i in range(오름.shape[0]):
  folium.Marker(
    [오름.iloc[i]['위도'], 오름.iloc[i]['경도']],
    popup=f'<div style="width:100px"><strong>{오름.iloc[i]["오름명"]}</strong><br>\
    주차장 : {오름.iloc[i]["주차장"]}<br>\
    화장실 : {오름.iloc[i]["화장실"]}<br>\
    <img width="80px" src="{img}"><br>\
    <a href="https://visitjeju.net/kr/search?q=%EC%98%A4%EB%A6%84">상세페이지 이동</a></div>',
    tooltip=tooltip
  ).add_to(m)

  folium.Circle(
      [오름.iloc[i]['위도'], 오름.iloc[i]['경도']],
      radius=오름.iloc[i]['표고(m)'],
      color='red',
      fill = False,
      tooltip=tooltip
  ).add_to(m)

  location_one= [33.3684955195788, 126.52918183373025]
  location_two=[오름.iloc[i]['위도'], 오름.iloc[i]['경도']]
  거리 = haversine(location_one, location_two,unit='km')

  folium.PolyLine(
    locations= [location_one, location_two],
    tooltip= f'한라산에서{오름.iloc[i]["오름명"]}까지 {거리 :.2f} km'
  ).add_to(m)

m.save('map.html')  # 지도를 HTML 파일로 저장

m
```

# 2. 아이콘, 그룹체크박스, 클러스터 설정하기

## 아이콘 기본 설정

```python
from folium.features import CustomIcon

m = folium.Map(
  location=[33.3684955195788, 126.52918183373025],    # 한라산 위경도
  zoom_start=10,
  #tiles ='Stamen Terrain' # 레이아웃형태
)

tooltip = 'Click!!'

for i in range(오름.shape[0]):
  icon_img= "https://drive.google.com/file/d/1xWV05hdOQG5Z4T2dHtlEMgkEcMyfj5Pm/view?usp=sharing"
  icon=CustomIcon(
      icon_img,
      icon_size=(20,20),
      popup_anchor=(30,-30),
  )

  folium.Marker(
    [오름.iloc[i]['위도'], 오름.iloc[i]['경도']],
    popup=f'<div style="width:100px"><strong>{오름.iloc[i]["오름명"]}</strong><br>\
    주차장 : {오름.iloc[i]["주차장"]}<br>\
    화장실 : {오름.iloc[i]["화장실"]}<br>',
    tooltip=tooltip,
    icon=icon
  ).add_to(m)

m
```

## 그룹 체크 박스

```python
from folium.features import CustomIcon

m = folium.Map(
  location=[33.3684955195788, 126.52918183373025],    # 한라산 위경도
  zoom_start=10,
  #tiles ='Stamen Terrain' # 레이아웃형태
)

tooltip = 'Click!!'

주차장유=folium.FeatureGroup(name='주차장유').add_to(m)
주차장무=folium.FeatureGroup(name='주차장무').add_to(m)

for i in range(오름.shape[0]):
  if 오름.iloc[i]["주차장"] =='Y':
    folium.Marker(
      [오름.iloc[i]['위도'], 오름.iloc[i]['경도']],
      popup=f'<div style="width:100px"><strong>{오름.iloc[i]["오름명"]}</strong><br>\
      주차장 : {오름.iloc[i]["주차장"]}<br>\
      화장실 : {오름.iloc[i]["화장실"]}<br>',
      tooltip=tooltip,
    ).add_to(주차장유)
  else:
    folium.Marker(
      [오름.iloc[i]['위도'], 오름.iloc[i]['경도']],
      popup=f'<div style="width:100px"><strong>{오름.iloc[i]["오름명"]}</strong><br>\
      주차장 : {오름.iloc[i]["주차장"]}<br>\
      화장실 : {오름.iloc[i]["화장실"]}<br>',
      tooltip=tooltip,
    ).add_to(주차장무)

folium.LayerControl(collapsed=False).add_to(m) # 우측상단에 선택창

m
```

## 사각형, 다각형 찍기

```python

m = folium.Map()
folium.Rectangle(
    bounds= [[60, 40],[50,40],[60,90],[50,90]],
    tooltip='사각형'
).add_to(m)


m
```

```python

m = folium.Map()
folium.Polygon(
    locations= [[50,40],[60, 40],[60,90],[50,90],[45,75]],
    tooltip='다각형'
).add_to(m)


m
```
